```python
import pandas as pd
```

```python
def fill_none(df: pd.DataFrame) -> None:
    for i in df.columns:
        if df[i].dtype in ['object', 'string']:
            df[i] = df[i].fillna(df[i].mode()[0] if not df[i].mode().empty else "Unknown")
        else:
            df[i] = df[i].fillna(df[i].median())
```
